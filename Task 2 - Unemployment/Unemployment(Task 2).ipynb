# Visulaization

```python
sns.histplot(df.iloc[:,3], kde=True)
plt.title('Distribution of Unemployment Rate')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.pairplot(df)
```

```python
fig = plt.figure(figsize = (10, 10))
sns.histplot(x=' Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
sns.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (30, 15))
sns.histplot(x=' Estimated Labour Participation Rate (%)', data=df, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
y=df[' Estimated Unemployment Rate (%)']
x=df['Region']
plt_1 = plt.figure()
plt.title('Umemployment Rate')
plt.xlabel("States")
plt.ylabel("Estimated Unemployment rate",)
plt.xticks(rotation='vertical',fontsize=12)
sns.histplot(x,)
```

```python
sns.histplot(x=' Estimated Employed', data=df, kde=True, hue='Area')
plt.title('Employment according to Area')
plt.xlabel('Employment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (9, 9))
sns.lineplot(y=' Estimated Employed', x=' Date', data=df)
plt.title('Employment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Employment Rate')
plt.show()
```

```python
encode=pd.get_dummies(df)
encode
plt.figure(figsize=(12, 10))
sns.heatmap(encode.corr())
plt.show()
```

```python
plt.style.use("dark_background")
data_grouped = df.groupby('Region')[' Estimated Unemployment Rate (%)'].sum()
data_grouped.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Estimated Unemployment Rate by Region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python

sns.boxplot(x='Region', y=' Estimated Unemployment Rate (%)', data=df)
plt.xlabel('Region')
plt.ylabel('Estimated Unemployment Rate')
plt.title('Box Plot of Estimated Unemployment Rate by Region')
plt.xticks(rotation=45)
plt.show()
```
